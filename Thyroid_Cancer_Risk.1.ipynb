```python
# Barplot
counts_of_risk = df['Risk_Score'].value_counts()
sns.barplot(x=counts_of_risk.index, y=counts_of_risk.values,palette='Set2',edgecolor = "black")
plt.title('Risk to get Thyroid cancer',fontweight='bold')
plt.ylabel('Count',fontweight='bold')
plt.show()
```

```python
sns.countplot(data=df, x='Thyroid_Cancer_Risk', hue='Family_History',palette='Set2',edgecolor = "black")
plt.title('Thyroid Cancer Risk by Family History',fontweight='bold')
plt.show()
```

 Main Observations:

Low Risk: Most individuals without a family history fall into this category. The green bar (no family history) is much taller than the orange one (with family history).

Medium Risk: The difference narrows. People with and without family history are more evenly distributed, but those without still slightly outnumber.

High Risk: The trend flipsâ€”more individuals with a family history are in the high-risk group than those without.

```python
plt.figure(figsize=(8,6))
sns.countplot(data=df, x='Ethnicity', hue='Thyroid_Cancer_Risk',palette='Set2',edgecolor = "black")
plt.title('Cancer Risk Levels Across Regions',fontweight='bold')
plt.show()
```

Key Observations:

Caucasian group has the highest overall counts across all risk levelsâ€”especially in the Low and Medium categories.

Asian and African groups show similar distributions, with a noticeable presence in the High risk category.

Hispanic individuals have moderate counts in Low and Medium risk levels, but fewer in High risk.

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='Country', hue='Thyroid_Cancer_Risk',palette='Set2',edgecolor = "black")
plt.title('Cancer Risk Levels Across Countries',fontweight='bold')
plt.xticks(rotation=45)
plt.show()
```

Key Insights:

USA and India show the highest overall counts, especially in the Low and Medium risk categories.

South Korea and Japan have relatively high counts in the High risk category, suggesting a greater proportion of individuals at serious risk.

Nigeria and Brazil display lower overall counts, with a more balanced distribution across all three risk levels.

Germany, UK, and Russia lean toward Medium risk dominance, while China has a fairly even spread.

What This Might Indicate:

Healthcare access and screening: Countries like the USA and India may have broader screening programs, leading to higher detection rates in Low and Medium categories.

Environmental or genetic factors: The elevated High risk counts in East Asian countries (South Korea, Japan) could point to regional health patterns worth investigating.

Population size and data representation: Larger countries may naturally show higher counts, but the proportions between risk levels still reveal meaningful trends.

```python
# Heatmap for Correlations
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True),annot=True, cmap="coolwarm",fmt =".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Nodule Size vs. Diagnosis
plt.figure(figsize=(8,5))
sns.violinplot(x="Diagnosis", y = "Nodule_Size", data=df , palette="pastel")
plt.title("Distribution of Nodule Size by diagnosis")
plt.show()
```

Key Takeaways:

Similar distributions: The shape of both violins is quite similar, meaning the spread and density of nodule sizes are comparable for benign and malignant cases.

Overlap in size: There's no clear separation in nodule size between the two categories. This suggests that nodule size alone may not be a reliable indicator for distinguishing between benign and malignant nodules.

Median and IQR: The central box plot inside each violin shows the median and interquartile range (IQR). If these are close between the two groups, it reinforces that the central tendency and variability are alike.

```python
coun = df.groupby('Country')['Gender'].value_counts(normalize = True).unstack().reset_index()
for index, row in coun.iterrows():
    # Get the gender columns (assumes first column is 'Country')
    gender_labels = row.index[1:]
    gender_values = row[gender_labels].values
    
    plt.figure(figsize=(6,6))
    plt.pie(gender_values, labels=gender_labels, autopct='%1.1f%%', startangle=90)
    plt.title(f"Gender Distribution in {row['Country']}")
    plt.axis('equal') 
    plt.show()
```

```python
# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(country_counts, 
        labels=country_counts.index, 
        autopct='%1.1f%%',  # Display percentage on the chart
        startangle=140)      # Starting angle for better visual layout

plt.title('Country Distribution')
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle
plt.show()
```
